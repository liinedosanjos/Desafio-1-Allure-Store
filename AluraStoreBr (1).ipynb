### Importação dos dados



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

loja = pd.read_csv(url)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

loja.head()
```

#1. Análise do faturamento


```python
faturamento_loja1 = loja['Preço'].sum()
faturamento_loja2 = loja2['Preço'].sum()
faturamento_loja3 = loja3['Preço'].sum()
faturamento_loja4 = loja4['Preço'].sum()
print(f'Faturamento Loja 1: R$ {faturamento_loja1:,.2f}')
print(f'Faturamento Loja 2: R$ {faturamento_loja2:,.2f}')
print(f'Faturamento Loja 3: R$ {faturamento_loja3:,.2f}')
print(f'Faturamento Loja 4: R$ {faturamento_loja4:,.2f}')
```

```python
import matplotlib.pyplot as plt

# Faturamento (substitua pelos seus dados reais se já calculou)
faturamentos = [
    faturamento_loja1,
    faturamento_loja2,
    faturamento_loja3,
    faturamento_loja4
]
lojas = ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4']

# Cores diferentes para cada barra
cores = ['skyblue', 'salmon', 'lightgreen', 'plum']

# Criar gráfico
plt.figure(figsize=(10, 6))
bars = plt.bar(lojas, faturamentos, color=cores)

# Adicionando os valores no topo das barras
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 5000, f'R$ {yval:,.2f}', ha='center', fontsize=10, fontweight='bold')

# Estilização
plt.title('Faturamento Total por Loja', fontsize=14, fontweight='bold')
plt.ylabel('Faturamento (R$)', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()

plt.show()
```

# 4. Produtos Mais e Menos Vendidos

```python
# Loja 1
vendas_produtos_loja1 = loja['Produto'].value_counts()
mais_vendido_loja1 = vendas_produtos_loja1.idxmax()
menos_vendido_loja1 = vendas_produtos_loja1.idxmin()

# Loja 2
vendas_produtos_loja2 = loja2['Produto'].value_counts()
mais_vendido_loja2 = vendas_produtos_loja2.idxmax()
menos_vendido_loja2 = vendas_produtos_loja2.idxmin()

# Loja 3
vendas_produtos_loja3 = loja3['Produto'].value_counts()
mais_vendido_loja3 = vendas_produtos_loja3.idxmax()
menos_vendido_loja3 = vendas_produtos_loja3.idxmin()

# Loja 4
vendas_produtos_loja4 = loja4['Produto'].value_counts()
mais_vendido_loja4 = vendas_produtos_loja4.idxmax()
menos_vendido_loja4 = vendas_produtos_loja4.idxmin()

print(f"Loja 1 - Mais vendido: {mais_vendido_loja1} | Menos vendido: {menos_vendido_loja1}")
print(f"Loja 2 - Mais vendido: {mais_vendido_loja2} | Menos vendido: {menos_vendido_loja2}")
print(f"Loja 3 - Mais vendido: {mais_vendido_loja3} | Menos vendido: {menos_vendido_loja3}")
print(f"Loja 4 - Mais vendido: {mais_vendido_loja4} | Menos vendido: {menos_vendido_loja4}")
```

```python
import matplotlib.pyplot as plt

# Reutilizar a função de mais/menos vendidos
def top_produtos(df, top=3):
    contagem = df['Produto'].value_counts()
    mais_vendidos = contagem.head(top)
    menos_vendidos = contagem.tail(top)
    return mais_vendidos, menos_vendidos

# Dados das lojas
lojas_dfs = [loja, loja2, loja3, loja4]
nomes_lojas = ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4']

# Plot com subplots (1 linha, 4 colunas)
fig, axs = plt.subplots(2, 2, figsize=(18, 10))  # 2x2 grid
axs = axs.flatten()  # simplifica indexação

for i, df in enumerate(lojas_dfs):
    mais, menos = top_produtos(df)

    # Dados para plotar
    produtos = list(mais.index) + list(menos.index)
    quantidades = list(mais.values) + list(menos.values)
    cores = ['steelblue'] * 3 + ['indianred'] * 3

    # Plot em cada subplot
    axs[i].barh(produtos, quantidades, color=cores)
    axs[i].set_title(f'{nomes_lojas[i]} - Top 3 Mais e Menos Vendidos')
    axs[i].invert_yaxis()
    axs[i].grid(axis='x', linestyle='--', alpha=0.5)

    # Mostrar valores ao lado
    for idx, qtd in enumerate(quantidades):
        axs[i].text(qtd + 0.5, idx, str(qtd), va='center', fontsize=9)

plt.tight_layout()
plt.suptitle('Comparação dos Produtos Mais e Menos Vendidos por Loja', fontsize=16, fontweight='bold', y=1.03)
plt.show()
```

# 5. Frete Médio por Loja


```python
# Loja 1
frete_medio_loja1 = loja['Frete'].mean()
print(f"Frete médio - Loja 1: {frete_medio_loja1:.2f}")

# Loja 2
frete_medio_loja2 = loja2['Frete'].mean()
print(f"Frete médio - Loja 2: {frete_medio_loja2:.2f}")

# Loja 3
frete_medio_loja3 = loja3['Frete'].mean()
print(f"Frete médio - Loja 3: {frete_medio_loja3:.2f}")

# Loja 4
frete_medio_loja4 = loja4['Frete'].mean()
print(f"Frete médio - Loja 4: {frete_medio_loja4:.2f}")
```

```python
import matplotlib.pyplot as plt

# Dados reais ou previamente calculados
fretes = [
    frete_medio_loja1,
    frete_medio_loja2,
    frete_medio_loja3,
    frete_medio_loja4
]
lojas = ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4']
cores = ['gold', 'lightcoral', 'lightskyblue', 'mediumseagreen']

# Criar gráfico
plt.figure(figsize=(10, 6))
bars = plt.bar(lojas, fretes, color=cores)

# Adicionando os valores no topo das barras
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 0.5, f'R$ {yval:.2f}', ha='center', fontsize=10, fontweight='bold')

# Estilização
plt.title('Custo Médio de Frete por Loja', fontsize=14, fontweight='bold')
plt.ylabel('Valor em R$', fontsize=12)
plt.ylim(0, max(fretes) + 5)  # Margem extra para não cortar o texto
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()

plt.show()
```

🔍**Análise por Métrica**

💰* Faturamento Total*

A **Loja 1** teve o **maior faturamento**, com 1.534.509,12 reais, enquanto a **Loja 4** teve o **menor**, com 1.384.497,58 reais. Essa diferença indica que a **Loja 4** tem o menor desempenho financeiro, mesmo com frete mais barato.

🏷️ *Categorias mais vendidas*

As quatro lojas vendem uma variedade parecida de categorias. No entanto:

**Loja 3** se destaca em "móveis" e "eletrônicos", categorias de alto valor.

**Loja 4** se destaca em "brinquedos" e "utilidades domésticas", com menor ticket médio.

**Loja 1 e Loja 2** têm boa distribuição, mas a Loja 1 perde em variedade e avaliações.


📦 *Produtos mais e menos vendidos*

A análise revelou que:

Loja 1 tem três produtos com alta demanda e poucos produtos encalhados.

Loja 3 tem equilíbrio entre mais e menos vendidos.

Loja 2 vende bem produtos educacionais e tem menor número de produtos encalhados.

Loja 4 tem os produtos menos vendidos com as piores quantidades entre todas as lojas, indicando possível problema de curadoria ou público-alvo mal definido.

🌟 Avaliação dos Clientes

A Loja 3 possui a melhor avaliação média (4.05), seguida da Loja 2 (4.04).
A Loja 1 teve a pior média (3.98), o que pode indicar insatisfação dos clientes apesar do faturamento alto.
A Loja 4 teve avaliação intermediária (4.00), mas sem destaque.

🚚 Frete Médio

A Loja 4 possui o frete médio mais barato (R$ 31,28), porém, esse fator não compensa seu baixo faturamento, menor giro de produtos e desempenho geral inferior.

✅ **Conclusão**

Com base nas análises de faturamento, avaliação dos clientes, produtos vendidos, categorias mais populares e custo de frete, **a recomendação é que o Senhor João venda a Loja 4.**

Embora tenha o frete mais barato e boas vendas em categorias de baixo valor, a **Loja 4**:

*   Tem o menor faturamento
*   Apresenta os piores produtos menos vendidos
*   Possui uma linha de produtos com menor valor agregado
*   Não se destaca em nenhuma métrica principal

A **Loja 3**, por outro lado, se mostra a mais sólida e equilibrada, sendo a que **deve ser mantida como referência de sucesso**.
